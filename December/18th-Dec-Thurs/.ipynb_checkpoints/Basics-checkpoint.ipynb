## Number Guessing Game

```python
secret_number = 56

guess = int(input("Enter the Guess : "))

if secret_number == guess :
    print("You Got it")
else:
    print("Try Again")
```

### For Loop Usage

#### usage : when the end is defined

```python
secreat_number=56
for i in range(7):
    guess = int(input("Enter the guess : "))
    if secreat_number == guess:
        print("you got it")
        break
    else:
        print("Try Again")
```

### While Loop

#### usage : when the end is not defined wants to run the loop at infinite

```python
secreat=78
while True:
    guess = int(input("Enter the guess : "))
    if secreat == guess:
        print("you got it")
        break
    else:
        print("Try Again")
```

#### Using Multiple Conditions

```python
print('''
***********************************
* Welcome to Number Guessing Game *
***********************************
''')
secreat=78
for i in range(1,8):
    guess = int(input("Enter your guess : "))
    if 1<=guess<=100:
        if secreat>guess:
            print(f"{guess} is less than secreat number")
        elif secreat<guess:
            print(f"{guess} is greater than secreat number")
        else:
            print(f"you got it in {i} {"attempt" if i<=1 else "attempts"}")
            break
    else:
        print("Invaild Number")
        break
```

#### try catch exception

```python
import random
secreat=random.randint(1,100)
i=0
while True:
    i=i+1
    try:
        guess = int(input("Enter your guess : "))
        if 1<=guess<=100:
            if secreat>guess:
                print(f"{guess} is less than secreat number")
            elif secreat<guess:
                print(f"{guess} is greater than secreat number")
            else:
                print(f"you got it in {i} {"attempt" if i<=1 else "attempts"}")
                break
        else:
            print("Invaild Number")
            break
    except ValueError:
        print("Invalid Charecter, Not Recognized")
```
