### List
- Duplicate Value
- hetrogenous data structure
- 1 Dimensional
- Indexing and Slicing is Possible
- Mutable (Updatable)

```python
cart=[]
```

```python
type(cart)
```

```python
items_to_buy= ['apple', 'orange','grapes','Pomogranate',23,34,67.2]
```

```python
for i in range(len(items_to_buy)):
    cart.append(items_to_buy[i])
```

```python
cart
```

```python
len(items_to_buy)
```

```python
cart[-1]
```

```python
cart[1:5]
```

```python
cart[-5:-1]
```

```python
cart[1]= 'watermelon'
```

```python
cart
```

### Sets
- 1 Dimensional
- Hetrogenous
- Does not allow duplicates
- Indexing and Slicing is not possible
- Mutable

```python
empty_cart=set()
```

```python
print(type(empty_cart))
```

```python
for i in range(len(items_to_buy)):
    empty_cart.add(items_to_buy[i])
```

```python
empty_cart
```

```python
empty_cart[0:2]
```

```python
empty_cart
```

```python
empty_cart.add("tomato")
```
