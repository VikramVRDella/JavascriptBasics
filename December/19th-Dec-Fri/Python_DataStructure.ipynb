#### Sets
* 1-Dimensional
* Hetrogenous
* Does not Allow Duplicate Values
* Indexing and Slicing is not accessible
* Mutable

```python
empty_cart_2=set() #initializing empty set
```

```python
empty_cart_2
```

```python
items_to_buy_2={"apple","orange","mango"} #Handling Data with set
```

```python
type(items_to_buy_2)
```

```python
items_to_buy_2.add("grapes")
```

```python
items_to_buy_2
```

```python
items_to_buy_2.discard("apple")
```
