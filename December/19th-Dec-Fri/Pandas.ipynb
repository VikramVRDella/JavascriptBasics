#### Finding the Items Sold in Restaurant

```python
hotels = pd.read_table("data.tsv.txt")
hotels
```

```python
hotels.nunique()
```

```python
print(hotels["item_name"].unique())
```

#### Finding the least and most sold item

```python
print(hotels["item_name"].unique())
```

```python
order=hotels.groupby(by='item_name')['quantity'].sum() #chain of methods
```

```python
top=order.sort_values(ascending=False)
```

```python
top.head()
```

#### Finding Least Values

```python
top=order.sort_values().head()
```

```python
top.head()
```
