### Bivarient Analysis
- Bar Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
cars=pd.read_csv("mtcars.csv")
cars
```

```python
sns.barplot(data=cars,x="cyl",y="mpg") #x - Discrete Data and y - Continuous Data
plt.title("Cylinders vs MPG")
plt.show()
```

```python
sns.scatterplot(data=cars,x="cyl",y="hp")
plt.show()
```
