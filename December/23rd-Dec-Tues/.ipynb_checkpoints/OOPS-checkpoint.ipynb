#### Class Creation

```python
class SecureUser:
    def __init__(self,user,password):
       self.user=user
       self.__password=password

    def get_password(self,attempt):
        n=4
        for i in range(1,n):
            if attempt == self.__password:
                return "Login Successfull"
                break
            else:
                n-=1
                return f"you have {n-1} attempts"
```

```python
print("User login")
user=input("login user : ")
password=int(input("password : "   ))
```

```python
log=SecureUser(user,password)
```

```python
log.get_password(2134)
```
