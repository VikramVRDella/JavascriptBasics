```python
from scipy import stats
```

#### The length of human pregnancy is normally distributed with a mean of 272 days and Standard deviation of 9
- 1. Find RV
- 2. Find the probability of pregnancy last more than 280 days
- 3. Find the probability of pregnancy last less than 250 days
- 4. Find the probablity of last between 265 and 280 days

#### Note
- **x = target that is -1SD and 1SD - 272-9 and 272+9**
- **loc = 272 the mean value**
- **scale = value of standard deviation**

#### RV - Human Pregnancy

#### 2nd

```python
(1-stats.norm.cdf(x=280,loc=272,scale=9))*100
```

#### 3rd

```python
(stats.norm.cdf(x=250,loc=272,scale=9))*100
```

#### 4th

```python
(stats.norm.cdf(x=280,loc=272,scale=9) - stats.norm.cdf(x=265,loc=272,scale=9))*100
```
