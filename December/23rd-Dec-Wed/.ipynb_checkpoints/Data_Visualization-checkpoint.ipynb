```python
import pandas as pd #Necessary Library Importing
```

```python
cars=pd.read_csv("mtcars.csv") #Importing Dataset
cars
```

### Data Understanding

#### Perform Data Visualization

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(cars["mpg"])
plt.xlabel("MPG")
plt.ylabel("frequency")
plt.title("MPG Distribution")
plt.show()
```

### Insights
In this given dataset 2 cars gives the least milage as 10 and the 2 cars atmost gives the milage at 30 and rest of the cars most on them will give the milage between 15-24 and consider buying it
