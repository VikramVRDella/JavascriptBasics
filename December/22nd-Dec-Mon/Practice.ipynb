#### Creating a list and printing the even Numbers only from the list

```python
numbers = range(1,51)
even_numbers = []
for i in numbers:
    if i%2 ==0:
        even_numbers.append(i)
print(even_numbers)
```

#### Counting Vowels in User Given word

```python
word = input("Enter a Word : ").lower()
vowels = 'aeiou'
count = 0

for i in word:
    if i in vowels:
        count+=1
print(count)
```

#### Reversing a list without using built-in function

```python
number = [1,2,3,4,5]
reverse = []
for i in number:
    reverse.append(number[-i])
print(reverse)
```
