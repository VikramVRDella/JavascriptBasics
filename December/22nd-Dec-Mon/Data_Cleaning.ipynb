### Data Cleaning

```python
import pandas as pd
```

```python
order_details = pd.read_csv("data.tsv.txt",sep="\t")
```

```python
order_details
```

```python
order_details.dtypes
```

```python
price=order_details["item_price"]
```

```python
item_update=price.str.replace("$","").astype(float)
```

```python
order_details["item_price"]=item_update
```

```python
top=order_details.groupby("item_name")["item_price"].sum().sort_values(ascending=False).head()
```
