#### Import Necessary Libraries

```python
import pandas as pd
```

#### Data Collection

```python
gender_classification = pd.read_csv("weight-height.csv")
gender_classification
```

#### Data Understanding
- Check Null items on dataset
- Check Datatypes and transform the dataset as your need
- Check the Shape

```python
gender_classification.shape
```

```python
gender_classification.isna().sum()
```

```python
gender_classification.dtypes
```

#### Data Preparation

```python
gender_classification['Gender'] = gender_classification['Gender'].replace(to_replace='Male',value=0)
```

```python
gender_classification['Gender'] = gender_classification['Gender'].replace(to_replace='Female',value=1)
```

```python
gender_classification
```

```python
gender_classification.dtypes
```

```python
gender_classification['Height'] = gender_classification['Height']*2.54 #converting Inches to cm
```

```python
gender_classification
```

```python
gender_classification['Weight'] = gender_classification['Weight']/2.205 #Converting pounds into kg
```

```python
gender_classification
```

#### Model Building

```python
x=gender_classification.drop('Gender',axis=1)
y=gender_classification.drop(['Height','Weight'],axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

#### Model Testing

```python
y_pred= model.predict(x)
```

```python
model.predict_proba(x)
```

#### Model Evaluation

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y,y_pred)
```

#### Model Deployment

```python
from pickle import dump
dump(model,open('Gender_Classification.pk1','wb'))
```

```python
from pickle import load
ml_model = load(open('Gender_Classification.pk1','rb'))
```

```python
def user_input():
    height=float(input("Enter the Height : "))
    weight=float(input("Enter thr Weight : "))
    data={"Height":height,"Weight":weight}
    features=pd.DataFrame(data,index=[0])
    return features

df = user_input()
predict_probabity = ml_model.predict_proba(df)
result="Female" if predict_probabity[0][1]>0.8 else "Male"
print(result)
print(predict_probabity)
```
