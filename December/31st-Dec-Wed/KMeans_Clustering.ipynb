#### Unsupervised Learning - KMeans Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
university = pd.read_csv('Universities.csv')
university
```

```python
university.drop('Univ',axis=1,inplace=True) #dropped the university Column
```

```python
university
```

```python
university.dtypes
```

```python
x=university.select_dtypes(include=['int64','float64'])
```

```python
scalar=StandardScaler()
```

```python
x_Scaled=scalar.fit_transform(x)
```

```python
x_Scaled
```

```python
model=KMeans(n_clusters=3,random_state=0)
```

```python
model.fit(x_Scaled)
```

```python
labels = model.labels_
```

```python
centroids = model.cluster_centers_
```

```python
print("Labels for Each Row : ",labels)
print("Centroids (Standardized) : ",centroids)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score
```
