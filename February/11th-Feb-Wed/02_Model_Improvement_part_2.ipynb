```python
import tensorflow as tf
```

```python
import numpy as np
```

```python
x = np.array([1., 3., 5., 7., 9., 11.,13.,15.], dtype="float32")
y = np.array([2., 4., 6., 8., 10., 12.,14.,16.], dtype="float32")
x = x.reshape(-1, 1)
```

```python
X= tf.cast(x,dtype=tf.float32)
Y= tf.cast(y,dtype=tf.float32)
X,Y
```

```python
# Basic Model
basic_model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])
```

```python
#model compiling
basic_model.compile(loss=tf.keras.losses.mae,optimizer=tf.keras.optimizers.Adam(),metrics=["mae"])
```

```python
# Fitting the Model
basic_model.fit(X,Y,epochs=10)
```

```python
y_pred = basic_model.predict(x)
```

```python
y_pred
```

```python
# Model with higher Number of layers compared to the basic model
```

```python
improved_model = tf.keras.Sequential([
    tf.keras.layers.Dense(100,activation="relu"),
    tf.keras.layers.Dense(1),
])
```

```python
#Compiling the Improved Model
improved_model.compile(loss=tf.keras.losses.mae,optimizer=tf.keras.optimizers.SGD(),metrics=["mae"])
```

```python
#Fitting Model
improved_model.fit(x,y,epochs=50)
```
