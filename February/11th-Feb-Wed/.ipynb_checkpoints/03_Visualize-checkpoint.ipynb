#### The 3 Sets on model training 
* Training Dataset - the model learns from it and it has the 75-80% of the data
* Validation Dataset - the model tuned on this data which typically 10-15% of the data
* Testing Dataset - the model gets evaluated on this data which is about 10-15% of the data
