```python
import tensorflow as tf
```

```python
x = tf.constant([
    [1],
    [3],
    [5],
    [7],
    [9]
])
```

```python
y = tf.constant([
    [2],
    [4],
    [6],
    [8],
    [10]
])
```

```python
x,y
```

```python
#Model Building
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(1))
```

```python
#Model Compilation
model.compile(loss=tf.keras.losses.mae,optimizer=tf.keras.optimizers.SGD(),metrics=["mae"])
```

```python
#Model Fitting
model.fit(x=x,y=y,epochs=25)
```

```python
y_pred = model.predict(x)
```

```python
y_pred
```

```python
#Finding Errors between y and y_pred
y_error3 = y_pred - y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(y,y_error3)
plt.show()
```

#### Steps to improve the Model
1. **Creating a Model** - here we want to add the multiple layers to the hidden layer or change activation function
2. **Compiling a Model** - Change the optimization function or learning rate
3. **Fitting Model** - More Epoches for the training model
