#### ***Major Modeling Steps in Tensorflow***

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
#creating features
X = np.arange(1,8,dtype=float)
y= np.arange(-8,-1,dtype=float)
```

***Turn Numpy Arrays  into Tensors***

```python
tensor1 = tf.cast(tf.constant(X),dtype=tf.float32)
tensor2= tf.cast(tf.constant(y),dtype=tf.float32)
tensor1,tensor2
```

```python
plt.scatter(tensor1,tensor2)
plt.title("Sample Data")
plt.show()
```

```python
#Creating Sequential Model API
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])
```

```python
#Compiling a Model
model.compile(
    loss=tf.keras.losses.mae,
    optimizer=tf.keras.optimizers.SGD(),
    metrics=["mae"]
)
```

```python
model.fit(tf.expand_dims(X, axis=-1), y, epochs=5)
```

```python
#Try prediction using builded model
y_pred = model.predict(tensor1)
y_pred
```

```python
error = y-y_pred
error
```

```python
plt.scatter(y_pred,y)
plt.title("Errors of the model")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
