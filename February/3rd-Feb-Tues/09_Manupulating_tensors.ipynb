#### Manupulating Tensors

```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
tensor1 = tf.constant([[1,2],[3,4],[5,6]])
tensor2 = tf.constant([[1,2,3],[4,5,6]])
```

```python
tensor1
```

```python
tensor2
```

```python
tensor_mod=tf.transpose(tensor1)
tensor_mod
```

```python
tensor_final_0=tf.tensordot(tf.reshape(tensor1,shape=(2,3)),tensor2,axes=0)
tensor_final_0
```

```python
tensor_final_1=tf.tensordot(tensor_mod,tensor2,axes=0)
tensor_final_1
```

```python
print({
    "Result through reshaping" :tensor_final_0,
    "Result through Transpose" :tensor_final_1
})
```

*Difference in both reshape and the transpose*

```python
tensor_X = tensor_final_1-tensor_final_0
tensor_X
```

```python
tensor_final_2 = tf.tensordot(tensor1,tf.transpose(tensor2),axes=0)
tensor_final_2
```

```python
tensor_final_3 = tf.tensordot(tensor1,tf.reshape(tensor2,(3,2)),axes=0)
tensor_final_3
```

```python
tensor_Y = tensor_final_3 - tensor_final_2
tensor_Y
```

```python
print({
    "x modified Tensor" : tensor_X,
    "y modified Tensor" : tensor_Y
})
```

*Generally transpose one of the tensor for satisfy the condition in it*
