#### *Squeezing Tensors*
- *These removes the extra 1-dimensional thing on the tensors*

```python
import tensorflow as tf
```

```python
tensor_1 = tf.constant(tf.random.uniform(shape=[50]),shape=(1,1,1,1,50))
tensor_1
```

```python
tensor_1.shape
```

```python
tensor_1_squeeze = tf.squeeze(tensor_1)
tensor_1_squeeze
```
