### *Math-Operations = `Squaring, log, & Square_roots`*

```python
import tensorflow as tf
```

```python
#Creating new Tensor
tensor_1 = tf.range(start=1,limit=11)
tensor_1
```

```python
#Square of the tensors
tf.square(tensor_1)
```

*Type-Casting for `tf.sqrt()` which is compact with float16,float32*

```python
tensor_fp = tf.cast(tensor_1,dtype=tf.float16)
tensor_fp
```

```python
# Tensor Square root
tf.sqrt(tensor_fp)
```

```python
#tensor log also supports only float
tf.math.log(tensor_fp)
```
