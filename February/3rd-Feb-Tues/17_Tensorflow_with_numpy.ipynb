#### ***Tensorflow and Numpy Compatability***

```python
import tensorflow as tf
import numpy as np
```

```python
# Creating a Tensor Array
tensor_1 = tf.constant(np.arange(10))
tensor_1
```

```python
import pprint
pprint.pp({
    "Tensor":tensor_1,
    "Numpy Array": np.array(tensor_1)
})
```

```python
tensor_np = np.array(tensor_1)
tensor_np,type(tensor_np)
```

```python
#converting numpy_arrays into tensors
np_tensor = tf.constant(tensor_np)
np_tensor.numpy()[8]
```

***Checking Each of the datatypes of numpy and the tensorflow's tensors***

```python
numpy_tensor = tf.constant(np.array([3.,4.,5.,6.,7.,8.]))
normal_tensor = tf.constant([3.,4.,5.,6.,7.,8.])
```

```python
#Checking dtypes of tensors
pprint.pp({
    "dtypes of numpy_tensor":numpy_tensor.dtype,
    "dtypes of normal_tensor":normal_tensor.dtype
})
```
