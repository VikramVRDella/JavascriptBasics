#### One-Hot *`Encoding`* Tensors
- *One-Hot-`Encoding` is also known as Numerical Encoding*

```python
import tensorflow as tf
```

```python
some_list = [0,1,2,3] # rgb like
tf.one_hot(some_list,depth=4) 
```

##### Customizing `on_value == 1` and `off_value == 0` in `tf.one_hot`

```python
tf.one_hot(some_list,depth=4,on_value="yes",off_value="no")
```
