### ***Introduction to Neural Networks with Tensorflow***
- Predicting a number (dependent variable) based on other number (independent variable)

```python
import tensorflow as tf
import numpy as np
import matplotlib as plt
```

```python
import pprint
pprint.pp({
    "Tensorflow Version":tf.__version__,
    "Numpy Version":np.__version__,
    "Matplotlib Version":plt.__version__
})
```

```python
#creating features
X = np.arange(1,8,dtype=float)
y= np.arange(-8,-1,dtype=float)
```

```python
#plot the data
plt.pyplot.scatter(X,y)
plt.pyplot.show()
```

```python
y == X - 9 # Neural Network want to learn
```
