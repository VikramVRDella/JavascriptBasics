***Type Casting on Tensorflow***

```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
tensor1 = tf.constant([[1,2],[3,4],[5,6]])
tensor2 = tf.constant([[1,2,3],[4,5,6]])
```

```python
tensor1
```

```python
tf.cast(tensor1,dtype=tf.float16)
```

```python
tf.cast(tensor2,dtype=tf.float32)
```

```python
tf.cast(tensor1,dtype=tf.int16)
```
