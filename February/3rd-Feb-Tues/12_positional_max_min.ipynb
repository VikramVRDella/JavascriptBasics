#### *Finding Positional Maximum and minimum* 

```python
import tensorflow as tf
```

```python
tf.random.set_seed(42)
tensor1 = tf.random.uniform(shape=[50])
tensor1
```

```python
# Positonal Maximum of tensor
position_max = tf.argmax(tensor1)
position_max
```

```python
#Value of Maximum
import pprint
max_value = tensor1[position_max]
pprint.pp({
    "Tensor_Value":max_value,
    "Numpy_Value": max_value.numpy()
})
```

```python
# Confirming it as True or False 
(max_value == tf.reduce_max(tensor1)).numpy()
```

```python
#Finding the positional minimum
positional_min = tf.argmin(tensor1)
positional_min
```

```python
min_value = tensor1[positional_min]
min_value
```

```python
tf.reduce_min(tensor1)
```

```python
(min_value == tf.reduce_min(tensor1)).numpy()
```

```python
#Value of Minimum
import pprint
max_value = tensor1[position_max]
pprint.pp({
    "Tensor_Value":min_value,
    "Numpy_Value": min_value.numpy()
})
```
