#### *Aggregation Types*
- Get the minimum
- Get the maximum
- Get the mean of the tensor
- Get the sum of the tensor

```python
import numpy as np
```

```python
tensor2 = tf.constant(np.random.randint(0,100,size = 50))
tensor2
```

```python
#size of the tensor
tf.size(tensor2)
```

```python
#shape of the tensor
tensor2.shape
```

```python
#Dimensions of tensor
tensor2.ndim
```

```python
#Minimum on the tensor
tf.reduce_min(tensor2)
```

```python
#Maximum of the tensor
tf.reduce_max(tensor2)
```

```python
#Mean of the Tensor
tf.reduce_mean(tensor2)
```

#### *Excercise : Finding the Tensor Standard Deviation and the Variance of tensor2*
- *Both supports only the dtype = float64* 
- *or use tensorflow_probability to make variance and std*

```python
tensor2_fp = tf.cast(tensor2,dtype=tf.float64)
tensor2_fp
```

```python
# Finding std deviation
tf.math.reduce_std(tensor2_fp)
```

```python
#Finding variance on tensor2_fp
tf.math.reduce_variance(tensor2_fp)
```
