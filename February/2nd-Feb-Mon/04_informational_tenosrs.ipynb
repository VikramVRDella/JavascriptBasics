### Getting Information from Tensors
- Rank
- Shape
- Size
- Dimension

```python
import tensorflow as tf
```

```python
#creating rank 4 tensor 
tensor4 = tf.ones(shape=(2,3,4,5))
tensor4
```

```python
tensor4[0]
```

```python
#shape
tensor4.shape
```

```python
#Dimensions
tensor4.ndim
```

```python
#size
tf.size(tensor4)
```

```python
print("Datatype of Tensors : ",tensor4.dtype)
print("Number of Dimensions (rank): ",tensor4.ndim)
print("Shape of Tensors : ",tensor4.shape)
print("Elements on 0 axis : ",tensor4.shape[0])
print("Elements on last axis : ",tensor4.shape[-1])
print("Total Number of Tensors : ",tf.size(tensor4))
print("Total Number of Tensors : ",tf.size(tensor4).numpy())
```
