#### Turn Numpy Arrays into tensors

```python
import numpy as np
np.__version__
```

```python
array1=np.arange(1,25,dtype="int32")
array1
```

```python
tensor3_without_shape=tf.constant(array1)
tensor3_without_shape
```

```python
tensor3_Third=tf.constant(array1,shape=(2,3,4))
tensor3_Third
```

```python
tensor3_Two=tf.constant(array1,shape=(6,4))
tensor3_Two
```

```python
tensor3_Two.ndim
```
