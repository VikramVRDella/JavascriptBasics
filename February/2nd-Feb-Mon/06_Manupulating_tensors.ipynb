### Manupulating Tensors

```python
import tensorflow as tf
```

#### Addition

```python
tensor1 = tf.constant([[10,7],[7,10]])
# tensor1 = tensor1+10
tensor1 + 10
```

```python
tensor1
```

#### Subraction

```python
tensor1 - 10
```

#### Multiplication

```python
tensor1 * 10
```

#### Division

```python
tensor1 / 10
```

#### Built-in Functions

```python
# Addition
print("tf.add")
print(tf.add(tensor1, 100))
#Subraction
print("tf.subract")
print(tf.subtract(tensor1, 10))
#Multiplication
print("tf.multiply")
print(tf.multiply(tensor1, 100))
#Division
print("tf.divide")
print(tf.divide(tensor1,5))
```

#### Matrix Multiplications

```python
tensor2 = tf.constant([[10,11],[12,14]])
tensor2
```

```python
tf.matmul(tensor1,tensor2)
```

```python
tensor1 @ tensor2
```
