#### Dot-Product(Matrix Multiplication)
- *Matrix Multiplication also referred as dot product*
- **tf.matmul()**
- **tf.tensordot()**

```python
import tensorflow as tf
```

```python
tensor1 = tf.constant([[1,2,3],[4,5,6]])
tensor2 = tf.constant([[1,2],[3,4],[5,6]])
```

```python
tensor1,tensor2
```

```python
tensor_mod = tf.reshape(tensor2,shape=(2,3))
tensor_mod
```

```python
tf.matmul(tensor1,tf.reshape(tensor2,shape=(2,3)),transpose_a=True)
```

```python
tensor1=tf.reshape(tensor1,(2,3))
tensor1
```

```python
tensormul = tf.tensordot(tf.reshape(tensor1,(2,3)),tensor2,axes=1)
tensormul
```
