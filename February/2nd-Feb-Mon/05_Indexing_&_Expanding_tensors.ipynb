#### Indexing Tensors
- Indexing can be happen like python lists

```python
tensor4 = tf.ones(shape=(2,3,4,5))
tensor4
```

```python
tensor4[:2,:2,:2,:2]
```

```python
tensor4[:1,:1,:1,:]
```

```python
tensor4[:1,:1,:,:]
```

```python
tensor4[:1,:1,:,:1]
```

```python
tensor2=tf.constant([
    [10,2],
    [7,10]
])
tensor2
```

```python
#last Item on the list
tensor2[-1]
```

```python
#upgrading rank 2 tensor into rank 3 tensor
tensor3 = tensor2[...,tf.newaxis]
tensor3
```

```python
#alternative to tf.newaxis
tensor3 = tf.expand_dims(tensor2,axis=-1)
tensor3
```

```python
tensor3 = tf.expand_dims(tensor2,axis=0)
tensor3
```
