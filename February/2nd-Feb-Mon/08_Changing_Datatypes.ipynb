#### Changing Datatypes of Tensors

```python
import tensorflow as tf
```

```python
tensor1= tf.constant([[1.7,2.7],[3.2,4.1]])
print(tensor1)
```

```python
tensor_mod = tf.cast(tensor1,dtype=tf.float16)
tensor_mod
```

```python
tf.cast(tensor_mod,dtype=tf.int32)
```

```python
tf.cast(tensor_mod,dtype=tf.float16)
```
