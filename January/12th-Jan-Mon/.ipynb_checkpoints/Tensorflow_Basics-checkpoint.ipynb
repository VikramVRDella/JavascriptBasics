#### Ranks in Tensorflow

```python
string2 = tf.Variable(["Hi","Hello"],tf.string)
string3 = tf.Variable(["Hi","Hello"],["Hi",["Hello","Edy"],"Hello"],tf.string)
string4 = tf.Variable(["Hi","Hello"],["Hello","Hi"],tf.string)
```

```python
print(string3)
```

```python
tf.rank(string3)
```

#### Changing Shape in Tensors

```python
tensor1 = tf.ones([2,3])
tensor2 = tf.reshape(tensor1, [3,2])
```

```python
print(tensor1)
```

```python
print(tensor2)
```

```python
tensor3 = tf.reshape(tensor2, [3,-1]) # negative value can't be reshaped so it returns the previous value
```

```python
print(tensor3)
```

```python
tensor5 = tf.ones([5,5,5,5])
print(tensor5)
```

#### Reshaping the above tensor

```python
tensor4 = tf.reshape(tensor5, [125,5])
print(tensor4)
```
